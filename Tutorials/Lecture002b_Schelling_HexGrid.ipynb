```
from IPython.display import HTML,SVG
import mola.renderP5JS as r2d
import random
import math
```

```
nc,nr = 40,40
states = [[random.randint(0,3) for x in range(nc)] for y in range(nr)]
```

```
print(states[:3][:3])
```

```
my_colors = [(200,200,200),(255,0,127),(0,127,255),(0,255,127)]
```

```
dimX = 10
dimY = math.sqrt(3)*dimX/2.0

r2d.beginDraw()

for r in range(nr):
  for c in range(nc):
    s = states[r][c]
    # shift odd rows right by half a cell
    offset_x = r%2 * dimX/2.0
    r2d.fill(my_colors[s])
    r2d.circle(offset_x+c*dimX,r*dimY,dimX)

HTML(r2d.endDraw())
```

```
# for a given cell coordinate (x,y) return a list of its 6 neighbours
# odd rows are shifted half a cell to the right
# topology = torus, wraps around borders top-bottom and left-right
def get_neighbours(x,y):
  neighbours = []
  if y%2==0: # even rows
    directions = [(1,0),(-1,0),(0,-1),( 1,-1),(0,1),( 1,1)]
  else: # odd rows
    directions = [(1,0),(-1,0),(0,-1),(-1,-1),(0,1),(-1,1)]
  for d in directions:
    ox = (x+d[0]) % nc
    oy = (y+d[1]) % nr
    neighbours.append(states[oy][ox])
  return neighbours
```

```
def swap_all(sts):
  movers = []
  new_states = [[v for v in sts[i]] for i in range(nr)]
  for r in range(nr):
      for c in range(nc):
        celltype = new_states[r][c]
        if celltype==0:
          continue
        nbs = get_neighbours(c,r)
        if nbs.count(celltype)+nbs.count(0)<4:
          movers.append((c,r))
          
  random.shuffle(movers)
  for pc,pr in movers:
    frees = [(c,r) for c in range(nc) for r in range(nr) if new_states[r][c]==0]
    random.shuffle(frees)
    cf,rf = frees[0]
    new_states[rf][cf] = new_states[pr][pc]
    new_states[pr][pc] = 0
  return new_states
```

```
for i in range(40):
  states = swap_all(states)
```

```
dimX = 10
dimY = math.sqrt(3)*dimX/2.0

r2d.beginDraw()

for r in range(nr):
  for c in range(nc):
    s = states[r][c]
    # shift odd rows right by half a cell
    offset_x = r%2 * dimX/2.0
    r2d.fill(my_colors[s])
    r2d.circle(offset_x+c*dimX,r*dimY,dimX)

HTML(r2d.endDraw())
```
