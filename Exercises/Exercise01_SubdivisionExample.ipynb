# Exercise: 16.384 Faces - Procedural Ornament

Create a colorful computational ornament based on shape grammars. Use maximal 16.384 faces.

  ## Submission 1: Procedure
Choose your prefered parameter setting and export each iteration of refinement as  OBJ File.

Filename Convention:** LastName_Procedure_00.obj** 


```
# place your code here

# create the initial mesh as starting point
mesh=mf.constructDodecahedron(0,0,0,5)
for i in range(3):
  # construct an empty mesh with Mesh()
  newMesh=Mesh()
  newMesh.faces=[]
  # apply multiple rules to refine the mesh
  for face in mesh.faces:
    resultingfaces=subdivision.extrudeToPointCenter(face,0.2)
    # extend is a python command to extend an existing list with a new list
    newMesh.faces.extend(resultingfaces)
  mesh=newMesh
  # colorize the mesh
  color.colorFacesByArea(mesh.faces)
  filename='dillenburger_Procedure'+str(i)
  io.exportOBJ(mesh,filename+'.obj')
  files.download(filename+'.obj')
  files.download(filename+'.obj.mtl')

# display the result in the browser
mesh.updateAdjacencies()
mesh=subdivision.subdivideCatmull(mesh)
color.colorFacesByCurvature(mesh.faces)
HTML(renderer3D.displayMesh(mesh))

# implement saving as obj
```

Filename Convention:** LastName_Matrix_00x00.obj** 


```
# place your code here

for y in range(4):
  row = ''
  for x in range(4):
    # generate mesh
    row = row+ str(x)+str(y)+' '
  print(row)
```

```
for x in range(4):
  for y in range(4):
    width=x+1
    depth=y+1
    mesh=mf.constructBox(0,0,0,width,depth,3)
    fileName='dillenburger_matrix_'+str(x)+'x'+str(y)
    io.exportOBJ(mesh,fileName+'.obj')
    print('filename: '+fileName)
```
