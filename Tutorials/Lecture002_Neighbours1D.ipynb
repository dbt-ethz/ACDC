```
import random 
import math
from IPython.display import HTML, SVG
import mola.renderP5JS as renderer2D
from mola.core import *
from mola import io
```

##Setup

```
#@title Define Preferences { run: "auto" }

preference_0_0 = -1 #@param [-1,0,1] {type:"raw"}
preference_0_1 = 1 #@param [-1,0,1] {type:"raw"}
preference_0_2 = 1 #@param [-1,0,1] {type:"raw"}
preference_1_1 = -1 #@param [-1,0,1] {type:"raw"}
preference_1_2 = 1 #@param [-1,0,1] {type:"raw"}
preference_2_2 = -1 #@param [-1,0,1] {type:"raw"}

preferences={}
preferences[(0,0)]=preference_0_0
preferences[(0,1)]=preference_0_1
preferences[(0,2)]=preference_0_2
preferences[(1,1)]=preference_1_1
preferences[(1,2)]=preference_1_2
preferences[(2,2)]=preference_2_2

def calculateFitness():
  sumFitness=0
  for i in range(len(cells)-1):
    v1=cells[i]
    v2=cells[i+1]
    if v1>v2:
      sumFitness+=preferences[(v2,v1)]
    else:
      sumFitness+=preferences[(v1,v2)]
  return sumFitness
            
def backup():
  global oldCells
  oldCells=list(cells)

def restore():
  global cells
  cells=list(oldCells)  

def swap():
  rIndex1=random.randrange(len(cells))
  rIndex2=random.randrange(len(cells))
  temp=cells[rIndex1]
  cells[rIndex1]= cells[rIndex2]
  cells[rIndex2]=temp
               
def mutateForBetter():
  global fitness
  backup()
  swap()
  newFitness=calculateFitness()
  if fitness>newFitness:
    restore()
  else:
    fitness=newFitness
      
```

```
# define initial state
dim=20
global cells
cells=[]
for i in range(32):
  cells.append(random.randrange(3))# randomly assigns 0 or 1 or 2
global fitness  
fitness=calculateFitness()
```

```
print("old fitness:" +str(fitness))
iterations = 500 #@param {type:"slider", min:1, max:500, step:1}
for i in range(iterations):
  mutateForBetter()
  
#display
renderer2D.beginDraw()
# draw relations
renderer2D.strokeWeight(2)
for i in range(0,len(cells)-1):
  fit=0
  v1=cells[i]
  v2=cells[i+1]
  if v1>v2:
    fit=preferences[(v2,v1)]
  else:
    fit=preferences[(v1,v2)]
  if fit==1:
    renderer2D.stroke(0,255,0)
  if fit==-1:
    renderer2D.stroke(255,0,0)
  if fit!=0:  
    x=i*dim+dim/2
    y=dim/2
    renderer2D.line(x,y,x+dim,y)
# draw cells
renderer2D.strokeWeight(2)
renderer2D.stroke(0,0,0)
for i in range(0,len(cells)):
  if cells[i]==0:
    renderer2D.fill(255,0,255)
  elif cells[i]==1:
    renderer2D.fill(0,255,255)
  elif cells[i]==2:
    renderer2D.fill(255,255,0)
  renderer2D.rect(i*dim+3,3,i*dim+dim-3,dim-3)

print("new fitness:" +str(fitness))
HTML(renderer2D.endDraw())
```
