```
from sklearn import datasets
from sklearn.svm import SVC
```

```
iris = datasets.load_iris()
print(iris.data.shape)
```

```
print(iris.target.shape)
```

```
import matplotlib.pyplot as plt
```

```
plt.scatter(iris.data[:,1],iris.data[:,3],c=iris.target,cmap='rainbow')
```

```
classifier = SVC(gamma='scale')
classifier.fit(iris.data[:-10], iris.target[:-10])
```

```
predictions = list(classifier.predict(iris.data[-10:]))
expected = iris.target[-10:]

for i in range(10):
  print('predicted: '+str(predictions[i])+' / expected: '+str(expected[i]))
```
